# I. Identification de la variable à expliquer et analyses

````python
cible = "cible_1"
print("Initial type of target column")
print(df[[cible]].dtypes)
#Target is a variable of type Object, transform to int.
#Difference of [[ ]] (to have dataframe/columns) from [] (to have list of element) 

print("Transformed type of target column")
df[cible]=df[cible].astype(int)
print(df[[cible]].dtypes)
````

````python
# Décoder les colonnes encodées en bytes (type objet)
def decode_bytes(val):
    if isinstance(val, bytes):  # Vérifie si la valeur est de type byte
        try:
            return val.decode('utf-8')  # Tente de décoder en 'utf-8'
        except UnicodeDecodeError:
            return val.decode('utf-8', errors='ignore')  # Ignore les erreurs de décodage
    return val

# Appliquer cette fonction à toutes les colonnes de type object
for col in df.select_dtypes(include=[object]):
    df[col] = df[col].map(decode_bytes)
````

````python
#Transformer le format de date :
print("Nombre de dates : ",df[['date']].nunique())
print("Format de date initial: ", df[['date']].dtypes)

df['date'] = pd.to_datetime(df['date'], format = '%Y%m')
print("Format de date final: ", df[['date']].dtypes)
````

````python
## Évolution de la proportion de défaut (variable cible = 1) dans le temps
tx_defaut = df.groupby(df['date'].dt.to_period('M'))[cible].mean() 
ylim_max = 0.1

import matplotlib.pyplot as plt 
# Tracer le taux de la variable cible au fil du temps
tx_defaut.plot(title="Proportion mensuelle de défaut sur une fenêtre d'un an")
plt.xlabel('Date')
plt.ylabel('Proportion de défaut dans le temps')
plt.ylim(0,ylim_max)
plt.show()
````

# II. Analyse exploratoire des variables explicatives continues

````python
from sklearn.model_selection import train_test_split

# 'Stratify' parameter makes a split so that the proportion of values in
# the sample produced will be the same as the proportion of values provided

# Might be interesting to add "sector" etc. 
df['stratify_param'] = df[cible].astype(str) + '_' + df['date'].astype(str)

# Split train-test stratifié en utilisant la colonne combinée
X_train, X_test, y_train, y_test = train_test_split(
    df,  # Les features incluant la date si elle est utilisée comme feature
    df[cible],                 # La variable cible
    stratify=df['stratify_param'],  # La nouvelle colonne pour la stratification
    test_size=0.3,                # La proportion de test
    random_state=42               # set.seed()
)

# Vous pouvez ensuite supprimer la colonne 'stratify_col' si elle n'est pas nécessaire pour l'entraînement
X_train = X_train.drop(['stratify_param'], axis=1)
X_test = X_test.drop(['stratify_param'], axis=1)
````

## II.2. Etude de liaison entre les variables quantitatives

Deux tests sont possibles :
- Le test de pearson (pour tester les liaisons affines)
- Le test de spearman (pour tester les laisons monotones)

````python
spearman_res=X_train[signif_conti_var].corr(method='spearman')
````

````python
import seaborn as sns
plt.figure(figsize=(12, 10))
sns.heatmap(spearman_res,cmap='coolwarm', linewidths=.5)

# Ajoutez un titre
plt.title('Matrice de Corrélation')

# Affichez le graphique
plt.show()
````

````python
# Pour chaque groupe, garder le couple le plus lié avec la variable cible

import numpy as np

kruskal_res.set_index('Columns', inplace=True)
# Suppose that spearman_res is your DataFrame containing the Spearman correlations
# Mask to keep only the upper triangle (excluding the diagonal)
mask = np.triu(np.ones(spearman_res.shape), k=1).astype(bool)
high_corr = spearman_res.where(mask)

# Filter for correlations > 0.7
high_corr_pairs = high_corr[abs(high_corr) > 0.7].stack().index

#set() only to have unique values
numerical_to_drop = set()
for var1, var2 in high_corr_pairs:
    if var1==var2:
        continue
    if abs(kruskal_res.loc[var1, 'Stat']) > abs(kruskal_res.loc[var2, 'Stat']):
        numerical_to_drop.add(var2)
    else:
        numerical_to_drop.add(var1)
   
    #if X_train[[var1,var2]].corr(method="spearman").iloc[1][0] >0.7:
     #   print("OK")

kruskal_res.reset_index('Columns', inplace=True)
numerical_to_drop
````

## II.4. Tests des valeurs extrêmes et discrétisation

````python
import matplotlib.pyplot as plt
import seaborn as sns

# En discrétisant les variables, nous n'avons plus de problèmes de outliers

# Supposons que 'numerical_columns' est une liste des colonnes numériques de votre DataFrame,
# 'X_train' est votre DataFrame contenant les données, et 'cible' est votre variable cible.

n_cols = 3

# Calculer le nombre de lignes nécessaires
n_rows = (len(numerical_columns) + n_cols - 1) // n_cols

# Créer des sous-graphiques
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))

# Aplatir la matrice d'axes pour faciliter l'itération
axes = axes.flatten()

# Itérer sur chaque colonne numérique et chaque axe
for i, col in enumerate(numerical_columns):
    # Tracer la densité pour chaque classe de la variable cible
    for target_class in X_train[cible].unique():
        sns.kdeplot(X_train[X_train[cible] == target_class][col], ax=axes[i], label=str(target_class), fill=True)
    
    axes[i].set_title(col)
    axes[i].legend(title='Cible')

# Si le nombre de graphiques est inférieur au nombre total de sous-graphiques,
# désactiver les axes restants
for j in range(i + 1, n_rows * n_cols):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
````
