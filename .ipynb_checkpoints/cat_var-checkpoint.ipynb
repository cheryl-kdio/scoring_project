## 1. Etude de liaison entre les variables catégorielles et la variable cible

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import math
from scipy import stats
import numpy as np

import importlib
import functions
importlib.reload(functions)
cramer_cible_res=functions.compute_cramers_v(X_train, object_columns,cible).sort_values(by = "Cramer_V", ascending=True).round(3)
````

````python
plt.barh(cramer_cible_res['Columns'], cramer_cible_res["Cramer_V"])

# Ajoutez des étiquettes et un titre
plt.xlabel('V de cramer')
plt.ylabel('Variable')
plt.title('Variables catégorielles dépendantes de la variable cible')
````

````python
signif_cat_var=(cramer_cible_res[cramer_cible_res['Cramer_V']>0.07]['Columns']).to_list()

len(signif_cat_var)
````

## 2. Etude de liaison entre les variables catégorielles

````python
importlib.reload(functions)
p_value_df,cramer_v_df=functions.stats_liaisons_var_quali(X_train,signif_cat_var)
np.fill_diagonal(cramer_v_df.values, 1)
````

````python
# Créer la heatmap
plt.figure(figsize=(12, 10))  # Vous pouvez ajuster la taille de la figure si nécessaire
sns.heatmap(cramer_v_df.astype(float), cmap='coolwarm', annot=True, fmt=".2f", linewidths=.5)
````

````python
# Filter for correlations > 0.7
importlib.reload(functions)
high_corr_pairs = functions.filter_correlated_variables(cramer_v_df,0.5)

cramer_cible_res.set_index('Columns', inplace=True)

#set() only to have unique values
cat_to_drop = set()
for var1, var2 in high_corr_pairs:
    if var1==var2:
        continue
    if abs(cramer_cible_res.loc[var1, 'Cramer_V']) > abs(cramer_cible_res.loc[var2, 'Cramer_V']):
        cat_to_drop.add(var2)
    else:
        cat_to_drop.add(var1)


cramer_cible_res.reset_index('Columns', inplace=True)
len(cat_to_drop)
````

````python
categorical_columns = set(signif_cat_var)-cat_to_drop
len(categorical_columns)
````
