## 1. Etude de liaison entre les variables quantitatives

Deux tests sont possibles pour tester la correlation entre deux variables quantitatives :
- Le test de pearson (pour tester les liaisons affines)
- Le test de spearman (pour tester les laisons monotones)

```python
spearman_res=X_train[signif_conti_var].corr(method='spearman')
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(spearman_res,cmap='coolwarm', linewidths=.5)

# Ajoutez un titre
plt.title('Matrice de Corrélation')

# Affichez le graphique
plt.show()
```

```python
# Pour les couples de variables correlés entre elles, ne garger que la variable le plus lié avec la variable cible (i.e. corr>0.7)
import numpy as np

kruskal_res.set_index('Columns', inplace=True)
# Suppose that spearman_res is your DataFrame containing the Spearman correlations
# Mask to keep only the upper triangle (excluding the diagonal)
mask = np.triu(np.ones(spearman_res.shape), k=1).astype(bool)
high_corr = spearman_res.where(mask)

# Filter for correlations > 0.7
high_corr_pairs = high_corr[abs(high_corr) > 0.7].stack().index

#set() only to have unique values
numerical_to_drop = set()
for var1, var2 in high_corr_pairs:
    if var1==var2:
        continue
    if abs(kruskal_res.loc[var1, 'Stat']) > abs(kruskal_res.loc[var2, 'Stat']):
        numerical_to_drop.add(var2)
    else:
        numerical_to_drop.add(var1)
   
    #if X_train[[var1,var2]].corr(method="spearman").iloc[1][0] >0.7:
     #   print("OK")

kruskal_res.reset_index('Columns', inplace=True)
numerical_to_drop
```

```python
#new size of numerical columns 
numerical_columns = set(signif_conti_var)-numerical_to_drop
len(numerical_columns)
```

## 3. Analyse des valeurs extrêmes et discrétisation

```python
import matplotlib.pyplot as plt
import seaborn as sns

# En discrétisant les variables, nous n'avons plus de problèmes de outliers

# Supposons que 'numerical_columns' est une liste des colonnes numériques de votre DataFrame,
# 'X_train' est votre DataFrame contenant les données, et 'cible' est votre variable cible.

n_cols = 3

# Calculer le nombre de lignes nécessaires
n_rows = (len(numerical_columns) + n_cols - 1) // n_cols

# Créer des sous-graphiques
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))

# Aplatir la matrice d'axes pour faciliter l'itération
axes = axes.flatten()

# Itérer sur chaque colonne numérique et chaque axe
for i, col in enumerate(numerical_columns):
    # Tracer la densité pour chaque classe de la variable cible
    for target_class in X_train[cible].unique():
        sns.kdeplot(X_train[X_train[cible] == target_class][col], ax=axes[i], label=str(target_class), fill=True)
    
    axes[i].set_title(col)
    axes[i].legend(title='Cible')

# Si le nombre de graphiques est inférieur au nombre total de sous-graphiques,
# désactiver les axes restants
for j in range(i + 1, n_rows * n_cols):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```
